```python
DJIA_stocks_df = pd.read_csv("data/upload_DJIA_table.csv")
Reddit_news_df = pd.read_csv("data/Combined_News_DJIA.csv")
```

```python
merged_df = pd.merge(Reddit_news_df,DJIA_stocks_df, on="Date" )
```

```python
#add dates as numeric
# merged_df['Date_num'] = merged_df['Date'].apply(lambda x: x.replace("-", ""))
# merged_df['Date_num'] = merged_df['Date_num'].apply(lambda x: pd.to_numeric(x))
# merged_df.sort_values('Date_num')

merged_df['Date_num'] = merged_df['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))


merged_df.head()
```

```python
merged_df.tail()
```

```python
fig, ax = plt.subplots(1, figsize = (16, 5))

dates = merged_df['Date_num'].to_list()
ax.plot(dates, merged_df['Close'])

ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.set_title('Close Price Over Time')
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=merged_df['Date_num'],
                open=merged_df['Open'],
                high=merged_df['High'],
                low=merged_df['Low'],
                close=merged_df['Close'])])

fig.update_layout(
    title='DJI Candlestick',
    yaxis_title='Price',
    shapes = [dict(
        x0='2010-08-08', x1='2015-07-01', y0=0, y1=1, xref='x', yref='paper',
        line_width=2)],
    annotations=[dict(
        x='2010-08-08', y=0.05, xref='x', yref='paper',
        showarrow=False, xanchor='left', text='Note Here')]
)

fig.show()
```
